```python
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('./data/opsim/runs_general.csv')
```

```python
df.shape
```

```python
df
```

```python
df_heatmap = pd.DataFrame(columns=['opsim', 'metric', 'value'])
```

```python
run_names = df['run_name'].values
```

```python
metrics = list( df.columns[1:21] )
```

```python
metrics
```

```python
df[metrics[0]].values.max()
```

```python
for opsim in run_names:
    for metric in metrics:
        value = df[df.run_name == opsim] [metric].values[0]
        
        row = {'opsim':opsim.replace('.', '_'), 'metric':metric, 'value': value / df[metric].values.max() }
        
        df_heatmap.loc[len(df_heatmap)] = row
        # print(row)
```

```python
df_heatmap.columns
```

```python
df_heatmap.fillna(0, inplace=True)
```

```python
list( np.unique(df_heatmap.opsim) )
```

```python
list( np.unique(df_heatmap.metric) )
```
